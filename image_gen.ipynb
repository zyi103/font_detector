```python
from essential_generators import DocumentGenerator
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path 
import random
```

```python
gen = DocumentGenerator()
p = Path(r'fonts').glob('**/*')
```

```python
# Settings
def create_png(fontname, sent, filename):
    W, H = (224, 224)
    print(fontname,sent,filename)
    # Font
    font = ImageFont.truetype(f"fonts/{fontname}.ttf", random.randint(20,24), encoding='utf-8')

    # Image
    image = Image.new("RGB", (W, H), "white")
    draw = ImageDraw.Draw(image)
    offset_w, offset_h = font.getoffset(sent)
    w, h = draw.textsize(sent, font=font)
    wOffset, hOffset = (random.uniform(0.1,1.9), random.uniform(0.1,1.9))
    pos = ((W-w-offset_w)/2*wOffset, (H-h-offset_h)/2*hOffset)

    # Draw
    draw.text(pos, sent, "black", font=font)

    # Save png file
    image.save(f"{ROOTFILE}/{fontname}/{filename}.png")
    
def create_sample_png(fontname):
    W, H = (224, 112)
    # Font
    font = ImageFont.truetype(f"fonts/{fontname}.ttf", 30, encoding='utf-8')

    # Image
    image = Image.new("RGB", (W, H), "white")
    draw = ImageDraw.Draw(image)
    offset_w, offset_h = font.getoffset(fontname)
    w, h = draw.textsize(fontname, font=font)
    pos = ((W-w-offset_w)/2, (H-h-offset_h)/2)

    # Draw
    draw.text(pos, fontname, "black", font=font)

    # Save png file
    image.save(f"sample/{fontname}.png")
```

```python
# generate sample
for font in p:
    if font.suffix == '.ttf':
        # test if folder exist
        font_name = font.stem
        create_sample_png(font_name)
```

```python
# generate pngs
ITERATION = 1000
ROOTFILE = 'data'

for font in p:
        if font.suffix == '.ttf':
        font_name = font.stem
        print(font_name)
        if not Path(f'{ROOTFILE}/{font_name}/').is_dir():
            Path(f'{ROOTFILE}/{font_name}/').mkdir(parents=True, exist_ok=True)
        for i in range(ITERATION):
            create_png(font_name, " ".join([gen.word() for i in range(3)]), i)
```
