```python
from essential_generators import DocumentGenerator
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path 
```

```python
gen = DocumentGenerator()
```

```python
# Settings
W, H = (224, 224)

def create_png(fontname, sent, filename):
    print(fontname,sent,filename)
    # Font
    font = ImageFont.truetype(f"fonts/{fontname}.ttf", 24, encoding='utf-8')

    # Image
    image = Image.new("RGB", (W, H), "white")
    draw = ImageDraw.Draw(image)
    offset_w, offset_h = font.getoffset(sent)
    w, h = draw.textsize(sent, font=font)
    pos = ((W-w-offset_w)/2, (H-h-offset_h)/2)

    # Draw
    draw.text(pos, sent, "black", font=font)

    # Save png file
    image.save(f"data/{fontname}/{filename}.png")
```

```python
# generate pngs
p = Path(r'fonts').glob('**/*')

for font in p:
    if font.suffix == '.ttf':
        # test if folder exist
        font_name = font.stem
        if not Path(f'data/{font_name}/').is_dir():
            Path(f'data/{font_name}/').mkdir(parents=True, exist_ok=True)
        for i in range(500):
            create_png(font_name, " ".join([gen.word() for i in range(3)]), i)
```
